```python
import numpy as np
from dt1 import *
```

```python
#Load the train data
M1 = np.genfromtxt('./monks-1.train', missing_values=0, skip_header=0, delimiter=',', dtype=int)
ytrn_1 = M1[:, 0]
Xtrn_1 = M1[:, 1:]

M2 = np.genfromtxt('./monks-2.train', missing_values=0, skip_header=0, delimiter=',', dtype=int)
ytrn_2 = M2[:, 0]
Xtrn_2 = M2[:, 1:]

M3 = np.genfromtxt('./monks-3.train', missing_values=0, skip_header=0, delimiter=',', dtype=int)
ytrn_3 = M3[:, 0]
Xtrn_3 = M3[:, 1:]


# Load the test data
N1 = np.genfromtxt('./monks-1.test', missing_values=0, skip_header=0, delimiter=',', dtype=int)
ytst_1 = N1[:, 0]
Xtst_1 = N1[:, 1:]

N2 = np.genfromtxt('./monks-2.test', missing_values=0, skip_header=0, delimiter=',', dtype=int)
ytst_2 = N2[:, 0]
Xtst_2 = N2[:, 1:]

N3 = np.genfromtxt('./monks-3.test', missing_values=0, skip_header=0, delimiter=',', dtype=int)
ytst_3 = N3[:, 0]
Xtst_3 = N3[:, 1:]
```

```python
depths=[i for i in range(1,11)]
```

```python
trn_err_1=[]
trn_err_2=[]
trn_err_3=[]

tst_err_1=[]
tst_err_2=[]
tst_err_3=[]

for i in depths:
    tree1 = id3(Xtrn_1, ytrn_1, max_depth=i)
    tree2 = id3(Xtrn_2, ytrn_2, max_depth=i)
    tree3 = id3(Xtrn_3, ytrn_3, max_depth=i)
    
    
    ypred_trn_1 = np.array([predict_example(x, tree1) for x in Xtrn_1])
    ypred_trn_2 = np.array([predict_example(x, tree2) for x in Xtrn_2])
    ypred_trn_3 = np.array([predict_example(x, tree3) for x in Xtrn_3])
    
    ypred_tst_1 = np.array([predict_example(x, tree1) for x in Xtst_1])
    ypred_tst_2 = np.array([predict_example(x, tree2) for x in Xtst_2])
    ypred_tst_3 = np.array([predict_example(x, tree3) for x in Xtst_3])
    
    trn_err_1.append(compute_error(ytrn_1, ypred_trn_1))
    trn_err_2.append(compute_error(ytrn_2, ypred_trn_2))
    trn_err_3.append(compute_error(ytrn_3, ypred_trn_3))
    
    tst_err_1.append(compute_error(ytst_1, ypred_tst_1))
    tst_err_2.append(compute_error(ytst_2, ypred_tst_2))
    tst_err_3.append(compute_error(ytst_3, ypred_tst_3))        
    
    
```

```python
import matplotlib.pyplot as plt          # For all our plotting needs
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))
plt.plot(depths, trn_err_1, linewidth=3)
plt.plot(depths, trn_err_2, linewidth=3)
plt.plot(depths, trn_err_3, linewidth=3)

plt.xlabel('Tree depth "d"', fontsize=16)
plt.ylabel('Train set error', fontsize=16)

plt.xticks(list(depths), fontsize=12)
plt.legend(['Monk-1', 'Monk-2','Monk-3'], fontsize=16,loc='upper left')
```

```python
plt.figure(figsize=(20,10))
plt.plot(depths, tst_err_1, linewidth=3)
plt.plot(depths, tst_err_2, linewidth=3)
plt.plot(depths, tst_err_3, linewidth=3)

plt.xlabel('Tree depth "d"', fontsize=16)
plt.ylabel('Test set error', fontsize=16)

plt.xticks(list(depths), fontsize=12)
plt.legend(['Monk-1','Monk-2','Monk-3'], fontsize=16,loc='upper left')
```
